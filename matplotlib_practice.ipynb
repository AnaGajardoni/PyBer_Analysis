```python
# Create the plot.
#plt.figure(figsize=(100,100))
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
plt.xlim("Jan", "Jun")
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston', markeredgewidth=2)
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
plt.xlim("Jan", "June")
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plot
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, color = "green", marker = "D", linewidth=2, label="Boston")

ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare(s)")
ax.set_ylim(0,45)
ax.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
fig, ax = plt.subplots()
#plt.plot(x_axis, y_axis, 'o')
ax.scatter(x_axis, y_axis, color="skyblue", s = [i * 5 for i in y_axis], alpha = .8, edgecolor="black", linewidth=2, label="Boston")
ax.legend()
ax.set_title("PyBer Fare By Month")
ax.set_ylabel("Date")
ax.set_xlabel("Fare ($)")
ax.set_ylim(0,45)
ax.invert_xaxis()
```

```python
# Module 5.1.8
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Using OO
fig, ax = plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

ax.pie(y_axis,labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Find two max values
x_axis.reverse()
y_axis.reverse()


aux_y_axis = list(y_axis)
aux_y_axis.sort(reverse = True)

ind1 = y_axis.index(aux_y_axis[0])
ind2 = y_axis.index(aux_y_axis[1])

explode_val = 12*[0]

explode_val[ind1] = .1
explode_val[ind2] = .1

fig, ax = plt.subplots(figsize=(8,8))


ax.pie(y_axis,labels=x_axis, autopct='%.1f%%', explode=explode_val, startangle=90, shadow=True, )
plt.show()
```
